```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel("cleaned_combined_fifa_wc (2).xlsx", engine="openpyxl")
print("✅ Cleaned dataset loaded successfully")
df.head()

```

```python
print("Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nData Info:")
print(df.info())

print("\nDescriptive Statistics:")
df.describe()
```

“This gives an overview of numeric columns, value ranges, and potential outliers.”

```python
missing = df.isnull().sum()
print("Missing Values:\n", missing[missing > 0])
```

“All missing values were already handled during Task 1 cleaning; none remain.”

```python
plt.figure(figsize=(8,4))
sns.histplot(df["win_rate"], bins=20, kde=True, color="teal")
plt.title("Distribution of Win Rate Across All Teams")
plt.xlabel("Win Rate"); plt.ylabel("Frequency")
plt.tight_layout(); plt.savefig("winrate_distribution.png")
plt.show()
```

“Most teams have win rates below 0.5, while finalist teams cluster above 0.7.”

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="goal_diff", y="points", hue="finalist", data=df, palette="coolwarm")
plt.title("Goal Difference vs. Points (Finalists Highlighted)")
plt.tight_layout(); plt.savefig("goal_diff_vs_points.png")
plt.show()
```

“Teams with higher goal difference and points have a much greater chance of reaching finals.”

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x="rank", y="win_rate", hue="finalist", data=df, palette="viridis_r")
plt.title("FIFA Rank vs. Win Rate (Lower Rank = Stronger Team)")
plt.tight_layout(); plt.savefig("rank_vs_winrate.png")
plt.show()
```

“Lower (better) FIFA rank corresponds to higher win rate, especially among finalists.”

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_df = df.select_dtypes(include=["number"])

# Now compute and visualize correlations
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.tight_layout()
plt.savefig("eda_correlation_heatmap.png")
plt.show()
```

goal_diff, win_rate, and points strongly correlate with being a finalist.

rank has a negative correlation (lower rank → higher success).

```python
plt.figure(figsize=(8,4))
df.groupby("association")["finalist"].sum().sort_values().plot(kind="barh", color="orange")
plt.title("Number of Finalists by Football Association")
plt.xlabel("Count"); plt.tight_layout()
plt.savefig("finalists_by_association.png")
plt.show()
```

“Europe (UEFA) and South America (CONMEBOL) dominate finals appearances.”
